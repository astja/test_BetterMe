```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import sqlalchemy
```

```python
sample = pd.read_csv('BM sample data.csv')

sample['ARPU'] = [3 if x =='ios' else 2 for x in sample['Platform']] 

sample['ARPU'] = np.where((sample['Country Code'] == 'UK') & (sample['Platform'] == 'android'), '3', sample['ARPU'])
sample['ARPU'] = np.where((sample['Country Code'] == 'UK') & (sample['Platform'] == 'ios'), '5', sample['ARPU'])

sample['ARPU'] = np.where((sample['Country Code'] == 'AU') & (sample['Platform'] == 'android'), '3', sample['ARPU'])
sample['ARPU'] = np.where((sample['Country Code'] == 'AU') & (sample['Platform'] == 'ios'), '5', sample['ARPU'])

sample['ARPU'] = np.where((sample['Country Code'] == 'US') & (sample['Platform'] == 'android'), '3', sample['ARPU'])
sample['ARPU'] = np.where((sample['Country Code'] == 'US') & (sample['Platform'] == 'ios'), '5', sample['ARPU'])

sample['ARPU'] = np.where((sample['Country Code'] == 'CA') & (sample['Platform'] == 'android'), '3', sample['ARPU'])
sample['ARPU'] = np.where((sample['Country Code'] == 'CA') & (sample['Platform'] == 'ios'), '5', sample['ARPU'])
sample.to_csv('sample.csv')
```
